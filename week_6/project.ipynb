### Neural Network Training on MNIST Dataset

In this project, we'll build and train a neural network using the **MNIST dataset**. The MNIST dataset consists of 28x28 grayscale images of handwritten digits (0-9). We'll go through the following steps:

1. **Load and Explore the Data**: Import the dataset and explore the sample data.
2. **Preprocess the Data**: Scale the images and split the dataset into training and test sets.
3. **Build the Model**: Create a simple neural network model using `MLPClassifier` from scikit-learn.
4. **Train the Model**: Train the model on the training data.
5. **Evaluate the Model**: Evaluate the model's performance on the test data.
6. **Visualize the Results**: Plot some sample digits and visualize the accuracy of the model.

#### Step 1: Load and Explore the MNIST Dataset

The MNIST dataset contains 70,000 images of handwritten digits (60,000 for training and 10,000 for testing). We will load the dataset and visualize some of the sample images to understand its structure.

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report, confusion_matrix
```

```python
# Step 1: Load the MNIST dataset
mnist = fetch_openml('mnist_784')
data = mnist['data']
labels = mnist['target']
```

```python
# Step 2: Visualize some sample digits
# Visualize the first 5 images in the dataset
fig, axes = plt.subplots(1, 5, figsize=(12, 4))
for i, ax in enumerate(axes):
    ax.imshow(data.iloc[i].values.reshape(28, 28), cmap='gray')
    ax.set_title(f"Label: {labels[i]}")
    ax.axis('off')
plt.show()
```

#### Step 2: Preprocess the Data

Before training, we need to normalize the data to ensure that all features are on a similar scale. We also split the dataset into training and testing subsets.

```python
# Step 2: Preprocess the Data
# Convert labels to integers
labels = labels.astype(int)

# Normalize the pixel values to range [0, 1]
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(data_scaled, labels, test_size=0.2, random_state=42)
```

#### Step 3: Build the Neural Network Model

We will use the **Multilayer Perceptron (MLP) classifier** from scikit-learn, which is a feedforward neural network that can be used for classification tasks.

```python
# Step 4: Build the Neural Network Model
# Create a neural network model with one hidden layer of 64 neurons
mlp = MLPClassifier(hidden_layer_sizes=(64,), max_iter=300, activation='relu', solver='adam', random_state=42)
```

#### Step 4: Train the Model

After building the model, we will train it using the training data and monitor its performance.

```python
# Step 4: Train the Model
mlp.fit(X_train, y_train)
```


#### Step 5: Evaluate the Model

Once the model is trained, we will evaluate its performance on the test data and calculate accuracy.


```python
# Step 5: Evaluate the Model
# Evaluate the model on the test data
y_pred = mlp.predict(X_test)
```

#### Step 6: Visualize the Results

We will plot the first few images of the test set, display their true labels, and compare them with the model's predicted labels to evaluate the performance visually.

```python
# Print the classification report
print("Classification Report:\n", classification_report(y_test, y_pred))

# Print the confusion matrix
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

### Neural Network Training on MNIST Dataset using TensorFlow

In this project, we'll build and train a neural network using the **MNIST dataset** with **TensorFlow**. The MNIST dataset consists of 28x28 grayscale images of handwritten digits (0-9). We'll go through the following steps:

1. **Load and Explore the Data**: Import the dataset and visualize the sample data.
2. **Preprocess the Data**: Scale the images and split the dataset into training and test sets.
3. **Build the Model**: Create a simple neural network model using TensorFlow.
4. **Train the Model**: Train the model on the training data.
5. **Evaluate the Model**: Evaluate its performance on the test data.
6. **Visualize the Results**: Plot some sample digits and visualize the predictions.


#### Step 1: Load and Explore the MNIST Dataset

The MNIST dataset consists of 60,000 training images and 10,000 test images of handwritten digits. We will load the dataset and visualize some sample images.

```python
# Import necessary libraries
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras import layers, models

# Step 1: Load the MNIST dataset
mnist = tf.keras.datasets.mnist
```

```python
# Load and split the dataset into training and testing sets
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

#### Step 2: Preprocess the Data

We need to normalize the images (pixel values between 0 and 1), and split the dataset into training and testing subsets.


```python
# Step 2: Visualize some sample digits
# Visualize the first 5 images in the dataset
fig, axes = plt.subplots(1, 5, figsize=(12, 4))
for i, ax in enumerate(axes):
    ax.imshow(X_train[i], cmap='gray')
    ax.set_title(f"Label: {y_train[i]}")
    ax.axis('off')
plt.show()
```

#### Step 3: Build the Neural Network Model

We will use **TensorFlow**'s Keras API to define a simple feedforward neural network (multilayer perceptron) with one hidden layer.

```python

# Normalize the pixel values to range [0, 1]
X_train = X_train / 255.0
X_test = X_test / 255.0

# Reshape the data to add a channel dimension for the grayscale images
X_train = X_train.reshape(-1, 28, 28, 1)
X_test = X_test.reshape(-1, 28, 28, 1)

model = models.Sequential([
    layers.InputLayer(input_shape=(28, 28, 1)),
    layers.Conv2D(32, (3, 3), activation='relu'),  # Convolutional layer
    layers.MaxPooling2D((2, 2)),  # Max pooling layer
    layers.Flatten(),  # Flatten the 2D data to 1D
    layers.Dense(64, activation='relu'),  # Hidden dense layer
    layers.Dense(10, activation='softmax')  # Output layer (10 classes)
])

# Compile the model with a loss function, optimizer, and evaluation metric
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

#### Step 4: Train the Model

We will train the model using the training dataset and monitor its performance during training.

```python
# Step 5: Train the Model
history = model.fit(X_train, y_train, epochs=5, validation_split=0.1)
```

#### Step 5: Evaluate the Model

After training, we will evaluate the model's performance on the test dataset.

```python
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")
```

#### Step 6: Visualize the Results

We will visualize the model's predictions for some test samples and compare them to the true labels.


```python
# Visualize the first 5 predictions of the test set
y_pred = model.predict(X_test)

fig, axes = plt.subplots(1, 5, figsize=(12, 4))
for i, ax in enumerate(axes):
    ax.imshow(X_test[i].reshape(28, 28), cmap='gray')
    ax.set_title(f"True: {y_test[i]} \nPred: {np.argmax(y_pred[i])}")
    ax.axis('off')
plt.show()

# Plot the training history (loss and accuracy)
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Accuracy over Epochs')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss over Epochs')
plt.legend()

plt.tight_layout()
plt.show()
```
