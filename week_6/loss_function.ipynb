# Loss Function and Gradient Descent

In this lesson, we’ll cover the **Loss Function** and **Gradient Descent** – two fundamental concepts in machine learning. The **Loss Function** measures how far off a model's predictions are from the actual values, and **Gradient Descent** is the algorithm that minimizes this loss by optimizing model parameters.

---

## Loss Function

The **Loss Function** quantifies the error between a model’s predictions and the actual data. By calculating this error, we know how well (or poorly) our model is performing.

### Purpose of the Loss Function

- **Quantifies Error**: Provides a numerical value representing the cost of incorrect predictions.
- **Optimization Objective**: The training process aims to minimize this error.
- **Guides Gradient Descent**: The gradient of the loss function helps determine the direction and amount of parameter updates.

### Common Types of Loss Functions

For different types of tasks, we use different loss functions:

1. **Mean Squared Error (MSE)**:
   - Commonly used for regression.
   - Formula:
     
     $\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$
    
   - \( y_i \): actual value, \( \hat{y}_i \): predicted value, \( n \): number of data points.

2. **Cross-Entropy Loss**:
   - Common in classification tasks.
   - Formula (for binary classification):
     
     $\text{Cross-Entropy} = -\frac{1}{n} \sum_{i=1}^{n} \left( y_i \log(\hat{y}_i) + (1 - y_i) \log(1 - \hat{y}_i) \right)$
     
   - This loss penalizes wrong predictions heavily, particularly when predictions are far from actual values.

### Visualization of the Mean Squared Error (MSE) Loss Function

Below, we plot the **Mean Squared Error (MSE)** function and show how it changes as predictions approach the true value.


```python
import numpy as np
import matplotlib.pyplot as plt

# Define the Mean Squared Error (MSE) Loss Function
def mse_loss(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

# Example data (true value and range of predictions)
y_true = 4  # true value
y_preds = np.linspace(-2, 10, 100)  # range of predictions

# Calculate loss for each prediction
loss_values = [mse_loss(y_true, y_pred) for y_pred in y_preds]

# Plotting the Loss Function
plt.figure(figsize=(10, 6))
plt.plot(y_preds, loss_values, color="purple", label=r'MSE Loss')
plt.xlabel("Predicted Value")
plt.ylabel("Loss (MSE)")
plt.title("Mean Squared Error Loss Function")
plt.legend()
plt.grid(True)
plt.show()
```
