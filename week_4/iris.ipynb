## IRIS

he Iris dataset is a popular dataset for exploring classification techniques, containing 150 samples from three species of Iris flowers (setosa, versicolor, and virginica). Each sample has four features: sepal length, sepal width, petal length, and petal width. Letâ€™s load and explore the Iris dataset using Python, Matplotlib, and Seaborn to gain insights.

#### step 1: Load and Preview the Dataset

The Iris dataset can be loaded directly from Seaborn's built-in datasets, making it easy to start exploring.

```python
# Import necessary libraries
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the Iris dataset
iris = sns.load_dataset('iris')

# Preview the dataset
print(iris.head())
```

```python
# Pair plot to visualize relationships between features
sns.pairplot(iris, hue='species', markers=["o", "s", "D"])
plt.suptitle("Pair Plot of Iris Dataset", y=1.02)
plt.show()
```

```python
df = iris.drop(columns=['species'])
corr_matrix = df.corr()

# Display the correlation matrix as a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Matrix of Iris Dataset")
plt.show()
```

```python
# KDE plot for each feature by species
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

plt.figure(figsize=(14, 10))
for i, feature in enumerate(features, 1):
    plt.subplot(2, 2, i)
    sns.kdeplot(data=iris, x=feature, hue="species", fill=True)
    plt.title(f"{feature.capitalize()} Distribution by Species")
plt.tight_layout()
plt.show()
```

```python
# Box plot for each feature by species
plt.figure(figsize=(14, 10))
for i, feature in enumerate(features, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(data=iris, x='species', y=feature)
    plt.title(f"{feature.capitalize()} Distribution by Species")
plt.tight_layout()
plt.show()
```

```python
# Violin plot for each feature by species
plt.figure(figsize=(14, 10))
for i, feature in enumerate(features, 1):
    plt.subplot(2, 2, i)
    sns.violinplot(data=iris, x='species', y=feature, palette="muted")
    plt.title(f"{feature.capitalize()} Distribution by Species")
plt.tight_layout()
plt.show()
```
