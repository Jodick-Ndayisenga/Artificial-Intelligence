#### 1. Overview of the Dataset

Before plotting, let's get a quick overview of our dataset:

```python
# Display the DataFrame
print(burundi_df.head())
print(burundi_df.describe())
print(burundi_df.info())
```

#### 2. Distribution of Total Income

Understanding the distribution of total income is crucial to assess income inequality within the population.

```python
# Distribution of Total Income
plt.figure(figsize=(10, 6))
sns.histplot(burundi_df['Total Income'], bins=10, color='blue', kde=True)
plt.title('Distribution of Total Income in Burundi')
plt.xlabel('Total Income ($)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

Why is this important?

1. This plot helps us see how total income is distributed across households.
2. If the distribution is skewed, it may indicate income inequality.
3. A clear view of the income spread can inform policy decisions and economic planning.

#### 3. Average Income by Source

Analyzing the average income by source (agricultural, business, employment, and remittances) can reveal which sectors contribute most to household income.

```python
# Average Income by Source
income_sources = burundi_df[['Agricultural Income', 'Business Income', 'Employment Income', 'Remittances']].mean()

plt.figure(figsize=(8, 6))
income_sources.plot(kind='bar', color=['green', 'orange', 'blue', 'purple'])
plt.title('Average Income by Source in Burundi')
plt.ylabel('Average Income ($)')
plt.xlabel('Source of Income')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

Why is this important?

1. Understanding income sources helps identify economic strengths and weaknesses in Burundi.
2. If a significant portion of income comes from agriculture, it suggests a need for policies that support this sector, while a diversified income may indicate economic resilience.

#### 4. Correlation Heatmap

Visualizing correlations between numerical variables helps understand relationships that might exist within the data.

```python
# Correlation Heatmap
plt.figure(figsize=(10, 6))
news_data = burundi_df.drop(["Region", "City"], axis = 1)
sns.heatmap(news_data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Income and Demographics')
plt.show()
```

Why is this important?

1. The heatmap reveals how different income sources relate to total income and household characteristics.
2. Strong correlations (e.g., between agricultural income and total income) indicate potential areas for investment or policy focus.

#### 12. Bar Plot of Average Income Sources by Region

This analysis can help us understand regional dependencies on different income sources.

```python
# Average Income by Source by Region
avg_income_by_region = burundi_df.groupby('Region')[['Agricultural Income', 'Business Income', 'Employment Income', 'Remittances']].mean()

plt.figure(figsize=(12, 6))
avg_income_by_region.plot(kind='bar', stacked=True)
plt.title('Average Income Sources by Region in Burundi')
plt.ylabel('Average Income ($)')
plt.xlabel('Region')
plt.xticks(rotation=45)
plt.legend(title='Income Source')
plt.grid(axis='y')
plt.show()
```

Why is this important?

1. his bar plot helps to visualize how income sources vary across regions, indicating economic strengths and vulnerabilities.
2. Insights can lead to targeted investments in specific income-generating sectors.

#### 13. Cumulative Distribution Function (CDF) of Total Income

Visualizing the CDF of total income helps us understand the proportion of households earning below certain income thresholds.

```python
# CDF of Total Income
plt.figure(figsize=(10, 6))
sns.ecdfplot(data=burundi_df, x='Total Income', color='blue')
plt.title('Cumulative Distribution Function of Total Income')
plt.xlabel('Total Income ($)')
plt.ylabel('CDF')
plt.grid(True)
plt.show()
```

Why is this important?

1. The CDF provides insight into income distribution and can help identify poverty levels.
2. It allows policymakers to understand how many households fall below certain income thresholds.

#### 14. Area Plot for Economic Overview

An area plot can visualize the cumulative contributions of different income sources to total income.

```python
# Area Plot for Economic Overview
plt.figure(figsize=(12, 6))
income_sources_cumulative = burundi_df[['Agricultural Income', 'Business Income', 'Employment Income', 'Remittances']].cumsum()
income_sources_cumulative.plot(kind='area', alpha=0.5)
plt.title('Cumulative Contributions of Income Sources')
plt.xlabel('Household Index')
plt.ylabel('Cumulative Income ($)')
plt.grid(axis='y')
plt.show()
```

Why is this important?

1. This plot shows how each income source contributes to total income over the dataset.
2. It helps visualize the impact of each sector on household economics.

#### 15. Pie Chart of Income Source Contributions

A pie chart can help visualize the proportion of total income derived from various sources.

```python
# Use the mean of each income source directly instead of summing them
income_sources = burundi_df[['Agricultural Income', 'Business Income', 'Employment Income', 'Remittances']].mean()

# Plot pie chart for the income sources
plt.figure(figsize=(8, 8))
plt.pie(income_sources, labels=income_sources.index, autopct='%1.1f%%', startangle=90, colors=['green', 'orange', 'blue', 'purple'])
plt.title('Proportion of Total Income by Source')
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
plt.show()
```

Why is this important?

1. This pie chart helps to quickly convey how dependent households are on different income sources.
2. It is particularly useful for stakeholders in determining where to focus economic development efforts.

#### 16. Violin Plot for Income Distribution by Region

Violin plots are useful for comparing distributions across multiple categories, providing insights into variance and probability density.

```python
# Violin Plot of Total Income by Region
plt.figure(figsize=(10, 6))
sns.violinplot(x='Region', y='Total Income', data=burundi_df, inner='quartile')
plt.title('Income Distribution by Region in Burundi')
plt.ylabel('Total Income ($)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

Why is this important?

1. Violin plots show the distribution shape, including density and outliers, for income across regions.
2. They provide a deeper understanding of economic disparities within regions.
