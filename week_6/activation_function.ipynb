# Activation Functions in Neural Networks

**Activation functions** are essential components in neural networks, determining the output of neurons by applying a mathematical transformation to the input. They introduce **non-linearity**, enabling networks to approximate complex functions and solve a wide range of problems.

---

## What is an Activation Function?

In a neural network:
- **Activation functions** decide whether a neuron should be activated or not based on the weighted sum of its inputs.
- They introduce **non-linearity**, allowing the network to learn complex patterns.

Without activation functions, the network would simply compute a linear transformation, making it unable to handle non-linear relationships.

### Why Non-Linearity Matters

- **Non-linearity** enables the network to learn from complex data with intricate patterns.
- Activation functions let the network stack layers to form complex decision boundaries, enabling deeper learning.

---

## Common Types of Activation Functions

Let's go over some widely-used activation functions, their properties, and their use cases.

### 1. Sigmoid Activation Function

The **Sigmoid** function squashes input values into a range between 0 and 1, making it suitable for binary classification.

**Formula**:

$\sigma(x) = \frac{1}{1 + e^{-x}}$

**Properties**:
- **Range**: $ (0, 1) $
- **Applications**: Often used in the output layer for binary classification.

### 2. ReLU (Rectified Linear Unit)

The **ReLU** function is one of the most popular activation functions for hidden layers in deep networks due to its simplicity and efficiency.

**Formula**:

$\text{ReLU}(x) = \max(0, x)$

**Properties**:
- **Range**: $ [0, \infty) $
- **Applications**: Common in hidden layers of deep networks.

### 3. Tanh (Hyperbolic Tangent) Function

The **Tanh** function is similar to the sigmoid but ranges between -1 and 1, making it useful for centered outputs.

**Formula**:

$\tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}$

**Properties**:
- **Range**: $ (-1, 1) $
- **Applications**: Often used in hidden layers, especially in early neural networks.

### 4. Softmax Activation Function

The **Softmax** function is typically used in the output layer of classification models, where it converts raw scores into probabilities.

**Formula**:

$\text{Softmax}(x_i) = \frac{e^{x_i}}{\sum_{j} e^{x_j}}$

**Properties**:
- **Range**: $ (0, 1) $
- **Applications**: Multi-class classification output layer.

---

## Choosing an Activation Function

The choice of an activation function depends on the task:

- **Sigmoid and Softmax** are popular for **output layers** in binary and multi-class classification, respectively.
- **ReLU** is commonly used in **hidden layers** of deep networks due to its simplicity and efficiency.
- **Tanh** can be effective in hidden layers when centered outputs (between -1 and 1) are desired.


```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Plotting the Sigmoid Function
x_vals = np.linspace(-10, 10, 100)
y_vals = sigmoid(x_vals)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label="Sigmoid", color="blue")
plt.title("Sigmoid Activation Function")
plt.xlabel("x")
plt.ylabel(r"$\sigma(x)$")
plt.legend()
plt.grid(True)
plt.show()
```

### **Neuron Network**

```python
def draw_neural_network(input_size, hidden_size, output_size):
    # Define coordinates for neurons in each layer
    layer_sizes = [input_size, hidden_size, output_size]
    v_spacing = 1  # vertical spacing between neurons
    h_spacing = 2  # horizontal spacing between layers
    
    # Set up the plot
    fig, ax = plt.subplots(figsize=(8, 6))
    ax.axis('off')
    
    # Calculate positions for each layer
    layer_x_positions = np.arange(len(layer_sizes)) * h_spacing
    max_layer_size = max(layer_sizes)
    
    # Draw neurons and connections
    for layer_idx, layer_size in enumerate(layer_sizes):
        # Vertical positions centered around max_layer_size
        y_positions = np.linspace(-v_spacing * (layer_size - 1) / 2, v_spacing * (layer_size - 1) / 2, layer_size)
        for neuron_idx in range(layer_size):
            x = layer_x_positions[layer_idx]
            y = y_positions[neuron_idx]
            # Draw neuron as a circle
            neuron = plt.Circle((x, y), radius=0.2, edgecolor='black', facecolor='lightblue', zorder=4)
            ax.add_patch(neuron)
            
            # Draw connections to next layer
            if layer_idx < len(layer_sizes) - 1:  # no connections from output layer
                next_y_positions = np.linspace(-v_spacing * (layer_sizes[layer_idx + 1] - 1) / 2,
                                               v_spacing * (layer_sizes[layer_idx + 1] - 1) / 2,
                                               layer_sizes[layer_idx + 1])
                for next_neuron_y in next_y_positions:
                    plt.plot([x, x + h_spacing], [y, next_neuron_y], 'k-', lw=0.5, color='gray')

    # Add layer labels
    plt.text(-0.5, max_layer_size * 0.5, "Input Layer", ha='center', fontsize=12, color='blue')
    plt.text(h_spacing, max_layer_size * 0.5, "Hidden Layer", ha='center', fontsize=12, color='blue')
    plt.text(2 * h_spacing, max_layer_size * 0.5, "Output Layer", ha='center', fontsize=12, color='blue')
    plt.show()

# Visualize a network with 3 input neurons, 4 hidden neurons, and 2 output neurons
draw_neural_network(input_size=3, hidden_size=4, output_size=2)
```
