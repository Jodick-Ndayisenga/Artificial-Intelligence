To create a complete course on data visualization using `matplotlib`, let's assume we have a sample dataset that contains the following information:

- Date: The day of the observation.
- Temperature: Daily temperature (in °C).
- Humidity: Daily humidity percentage.
- Wind Speed: Wind speed (in km/h).
- Weather Condition: Categories such as "Sunny," "Rainy," "Cloudy," etc.

The dataset will cover one month (30 days) and include variations that make it suitable for a range of plots. Here’s the data for the course, which we'll store in a pandas DataFrame and use to teach different plotting techniques.

#### Sample Data (Using a DataFrame)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Sample dataset
data = {
    'Date': pd.date_range(start='2023-01-01', periods=30),
    'Temperature': np.random.normal(25, 5, 30).round(1),  # Normally distributed around 25°C
    'Humidity': np.random.randint(40, 90, 30),  # Random humidity values between 40 and 90%
    'Wind Speed': np.random.randint(10, 40, 30),  # Random wind speeds between 10 and 40 km/h
    'Weather Condition': np.random.choice(['Sunny', 'Rainy', 'Cloudy', 'Windy'], 30)
}
df = pd.DataFrame(data)
df.head()
```

This D`ataFrame` `df` will be used to create various plots, teaching key `matplotlib` functions along the way.

#### 1. Line Plot: Temperature over Time

- Show trends over a time period.
- Learn to customize line style, color, and markers.

```python
plt.figure(figsize=(10, 5))
plt.plot(df['Date'], df['Temperature'], color='orange', marker='o', linestyle='-')
plt.title("Temperature Over Time")
plt.xlabel("Date")
plt.ylabel("Temperature (°C)")
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

#### 2. Scatter Plot: Temperature vs. Humidity

- Visualize relationships or correlations between two continuous variables.
- Experiment with color and size to add data dimensions.

```python
plt.figure(figsize=(10, 5))
plt.scatter(df['Temperature'], df['Humidity'], c='blue', alpha=0.7)
plt.title("Temperature vs Humidity")
plt.xlabel("Temperature (°C)")
plt.ylabel("Humidity (%)")
plt.show()
```

#### 3. Bar Chart: Average Temperature per Weather Condition

- Show categorical comparisons.
- Use groupby to aggregate data and plot categories.

```python
# Calculate average temperature per condition
avg_temp = df.groupby('Weather Condition')['Temperature'].mean()

plt.figure(figsize=(8, 5))
avg_temp.plot(kind='bar', color=['skyblue', 'lightgreen', 'salmon', 'lightgray'])
plt.title("Average Temperature by Weather Condition")
plt.xlabel("Weather Condition")
plt.ylabel("Average Temperature (°C)")
plt.show()
```

#### 4. Histogram: Distribution of Temperature

- Examine the distribution of a single variable.
- Adjust bins and explore styling.

```python
plt.figure(figsize=(8, 5))
plt.hist(df['Temperature'], bins=20, color='purple', edgecolor='orange')
plt.title("Distribution of Temperature")
plt.xlabel("Temperature (°C)")
plt.ylabel("Frequency")
plt.show()
```

#### 5. Pie Chart: Proportion of Weather Conditions

- Display parts of a whole for categorical data.
- Use autopct to show percentages.

```python
# Count occurrences of each weather condition
weather_counts = df['Weather Condition'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(weather_counts, labels=weather_counts.index, autopct='%1.1f%%', colors=['gold', 'lightblue', 'blue', 'red'])
plt.title("Weather Condition Distribution")
plt.show()
```

#### 6. Box Plot: Temperature Distribution by Weather Condition

- Understand the spread and detect outliers for a categorical variable.
- Plot multiple box plots for comparison.

```python
plt.figure(figsize=(10, 6))
df.boxplot(column='Temperature', by='Weather Condition', grid=False, patch_artist=True)
plt.title("Temperature Distribution by Weather Condition")
plt.suptitle("")  # Removes automatic title
plt.xlabel("Weather Condition")
plt.ylabel("Temperature (°C)")
plt.show()
```

#### 7. Stacked Bar Chart: Temperature and Humidity for Each Day

- Show the cumulative totals or parts of a whole.
- Useful for visualizing how different components contribute to a total.

```python
# Stacked bar chart
plt.figure(figsize=(10, 5))
plt.bar(df['Date'], df['Temperature'], label='Temperature (°C)', color='orange')
plt.bar(df['Date'], df['Humidity'], bottom=df['Temperature'], label='Humidity (%)', color='blue', alpha=0.5)
plt.title("Temperature and Humidity per Day")
plt.xlabel("Date")
plt.ylabel("Values")
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

### 8. Heatmap: Correlation Matrix

- Visualize correlations between multiple numerical features.
- Great for finding relationships between variables.

```python
import seaborn as sns

# Calculate correlation matrix
corr = df[['Temperature', 'Humidity', 'Wind Speed']].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

#### 9. Violin Plot: Temperature Distribution by Weather Condition

- Combine box plot with density plot to show data distribution by categories.
- Useful for visualizing the spread and concentration of data.

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='Weather Condition', y='Temperature', data=df, palette='muted')
plt.title("Temperature Distribution by Weather Condition")
plt.xlabel("Weather Condition")
plt.ylabel("Temperature (°C)")
plt.show()
```

#### 10. Line Plot with Multiple Lines: Temperature and Humidity Over Time

- Show multiple trends on the same plot.
- Use different colors and labels for each line.

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Temperature'], color='red', marker='o', linestyle='-', label='Temperature (°C)')
plt.plot(df['Date'], df['Humidity'], color='blue', marker='s', linestyle='--', label='Humidity (%)')
plt.title("Temperature and Humidity Over Time")
plt.xlabel("Date")
plt.ylabel("Values")
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

#### 11. 3D Plot: Temperature, Humidity, and Wind Speed

- Visualize three dimensions in a single plot.
- Suitable for complex, multi-dimensional data.

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

# 3D scatter plot
ax.scatter(df['Temperature'], df['Humidity'], df['Wind Speed'], c='red', marker='o')
ax.set_title("3D Plot of Temperature, Humidity, and Wind Speed")
ax.set_xlabel("Temperature (°C)")
ax.set_ylabel("Humidity (%)")
ax.set_zlabel("Wind Speed (km/h)")
plt.show()
```

#### 12. Annotations and Adding Text

Learn to add text and annotate data points.

```python
plt.figure(figsize=(10, 5))
plt.plot(df['Date'], df['Temperature'], color='orange', marker='o', linestyle='-')

# Annotate a point (e.g., the maximum temperature)
max_temp_day = df.loc[df['Temperature'].idxmax()]
plt.annotate(f"Max Temp {max_temp_day['Temperature']}°C", xy=(max_temp_day['Date'], max_temp_day['Temperature']),
             xytext=(max_temp_day['Date'], max_temp_day['Temperature'] + 5),
             arrowprops=dict(facecolor='black', arrowstyle="->"))

plt.title("Temperature Over Time with Annotation")
plt.xlabel("Date")
plt.ylabel("Temperature (°C)")
plt.xticks(rotation=45)
plt.grid()
plt.show()
```
